```python

cat_cols = ["city", "cuisine"]
num_cols = ["rating", "rating_count", "cost"]
```

```python

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from scipy.sparse import csr_matrix, hstack
import pickle

```

```python
# Scale numerical
scaler = StandardScaler()
scaled_num = scaler.fit_transform(c_df[num_cols])
num_sparse = csr_matrix(scaled_num)
```

```python
# Encode categorical
encoder = OneHotEncoder(
    handle_unknown="ignore",
    sparse_output=True
)

cat_sparse = encoder.fit_transform(c_df[cat_cols])
```

```python
# Final encoded matrix
encoded_data = hstack((num_sparse, cat_sparse)).tocsr()
```

```python
# Save objects
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

with open("encoder.pkl", "wb") as f:
    pickle.dump(encoder, f)
```

```python
##(COSINE SIMILARITY)
```

```python
user_input = {
    "rating": 4.2,
    "rating_count": 100,
    "cost": 250,
    "city": "Abohar",
    "cuisine": "Fast Food"
}

user_df = pd.DataFrame([user_input])
```

```python
###Transform User Input (NO FIT)
# Scale numerical
user_scaled_num = scaler.transform(user_df[num_cols])
user_num_sparse = csr_matrix(user_scaled_num)
```

```python

# Encode categorical
user_cat_encoded = encoder.transform(user_df[cat_cols])
```

```python
# Combine
user_vector = hstack((user_num_sparse, user_cat_encoded)).tocsr()
```

```python
##Compute Similarity (SAFE)
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(user_vector, encoded_data)

```

```python
##Top-N Recommendations
TOP_N = 10

top_indices = np.argsort(similarities[0])[::-1][:TOP_N]

recommendations = c_df.iloc[top_indices][
    ["name", "city", "rating", "rating_count", "cost", "cuisine"]
]

recommendations
```
